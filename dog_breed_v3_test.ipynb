```python
from keras.applications.efficientnet_v2 import preprocess_input
from keras.models import load_model
from keras.preprocessing.image import ImageDataGenerator
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import os
```

```python
labels_data = pd.read_csv('dpl-project-1-dog-breed-identification/labels.csv')
le = LabelEncoder()
le.fit(labels_data['breed'])

path = 'dpl-project-1-dog-breed-identification/test/'
img_h = 400
img_w = 400
img_c = 3

# my_model_v3_fine_tune.h5,
# my_model_v3_final.h5
# my_model_v3.h5
model = load_model('my_model_v3_final.h5')
```

```python
test_gen = ImageDataGenerator(preprocessing_function=preprocess_input)

data_test = test_gen.flow_from_directory(path,
                                            target_size=(img_h, img_w),
                                            batch_size= 64 ,shuffle=False,
                                            class_mode= 'binary',classes=['.'])
```

```python
results = model.predict(data_test, verbose='auto')
```

```python
Id = [x.split('.')[0] for x in os.listdir(path)]
```

```python
submission = pd.DataFrame(results, index=Id, columns=le.classes_)
submission.reset_index(inplace=True)
submission.rename(columns={'index':'id'}, inplace=True)
submission.to_csv('submission.csv', index=False)
```
